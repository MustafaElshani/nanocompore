### Generate a small dataset with both modified and unmodified conditions

```python
# Options
fasta = "./reference/ref.fa"
nreads_per_ref = 50

# Create directory for results
data_dir = "./sample_files/"

for replicate, nreads in [(1, 55), (2, 45)]:
    # Generate non modified control
    SimReads (
        fasta_fn=fasta,
        outpath=data_dir,
        prefix=f"unmodified_rep_{replicate}",
        nreads_per_ref= nreads)
    
    # Generate modified control
    SimReads (
        fasta_fn=fasta,
        outpath=data_dir,
        prefix=f"modified_rep_{replicate}",
        nreads_per_ref= nreads,
        intensity_mod_loc=10,
        dwell_mod_loc=0.01,
        mod_reads_freq=0.9,
        mod_bases_freq = 0.25,
        pos_rand_seed=2)
```
